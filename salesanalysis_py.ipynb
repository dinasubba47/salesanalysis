```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("sales_data.csv.xlsx")
print(data)
```

```python
print("first 5 rows of dataset:")
print(data.head())
```

```python
print(data.info())
```

```python
totalsales = data["sales"].sum()
print("\n Total sales",totalsales)
```

```python
#sales by product
productsales = data.groupby("product")["sales"].sum()
print("sales by product:")
print(productsales)
```

```python
#bar chart for product sales
productsales.plot(kind = "bar")
plt.title("sales by product")
plt.xlabel("product")
plt.ylabel("Total sales")
plt.show()
```
